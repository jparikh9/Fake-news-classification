```python
import pandas as panda
import nltk
```

```python
!pwd
#reading train data
train_data = panda.read_csv("train.csv", index_col='id')
print(train_data.shape)
train_data.head()
```

# New Section

```python
#dropping null values from tain.csv
train_data = train_data.dropna()
train_data.shape
```

```python
nltk.download('all')
```

```python
import string
from string import punctuation
from nltk.corpus import stopwords

# stopwords
stopwords_list = nltk.corpus.stopwords.words("english")
print(stopwords_list)
#making a list of stopwords and punctuation
stoplist = set(stopwords.words('english') + list(punctuation))
print(stoplist)
```

```python
#converting into lower case
train_data["title1_en"] = train_data["title1_en"].apply(lambda train_word: train_word.lower())
train_data["title2_en"] = train_data["title2_en"].apply(lambda train_word: train_word.lower())
train_data
```

```python
#removing stopwords from columns
train_data["title1_en"] = train_data["title1_en"].apply(lambda train_word: ' '.join(train_word for train_word in train_word.split() if train_word not in stoplist))
train_data["title2_en"] = train_data["title2_en"].apply(lambda train_word: ' '.join(train_word for train_word in train_word.split() if train_word not in stoplist))
train_data
```

```python
#removing punctuation
train_data["title1_en"] = train_data["title1_en"].apply(lambda train_word: ''.join(train_word for train_word in str(train_word) if train_word not in (punctuation)))
train_data["title2_en"] = train_data["title2_en"].apply(lambda train_word: ''.join(train_word for train_word in str(train_word) if train_word not in (punctuation)))
train_data
```

```python
#lemmatizing training data
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
porter = nltk.stem.PorterStemmer()
lemmatizer = nltk.stem.WordNetLemmatizer()
def lemmatize_text(text):
    lemma_words = [lemmatizer.lemmatize(word) for word in w_tokenizer.tokenize(text)]
    return " ".join(lemma_words)

train_data["title1_en"] = train_data["title1_en"].apply(lambda data: lemmatize_text(data))
train_data["title2_en"] = train_data["title2_en"].apply(lambda data: lemmatize_text(data))
train_data
```

```python
#storing preprocessed data in a csv file
from google.colab import files
train_data.to_csv('train_preprocessed.csv', encoding = 'utf-8-sig') 
files.download('train_preprocessed.csv')
```
